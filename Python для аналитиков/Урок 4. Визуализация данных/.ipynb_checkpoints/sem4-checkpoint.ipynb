### 2.3 Изучите распределение операционной системы

```python
data2 = df['OpSys'].value_counts()
data2
```

```python
plt.figure(figsize=(10, 10))
expl = [0.1, 0.03, 0.04, 0.05, 0.07, 0.06, 0.1, 0.1, 0.1]
plt.pie(data2.values, autopct='%1.1f%%', labels=data2.index, explode=expl)
# explode=explt - формирует размер "пирога" в зависимости от кол-ва элементов
plt.title('Доля типов носителей')
plt.legend(data2.index, loc='best');
#loc='best' - легенда ставится на оптимальное место, там где больше место
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x=data2.index, y=data2.values, palette='viridis')
# palette='' - цвет столбцов

plt.title('Анализ ноутбуков в зависимости от операционной системы')
plt.xlabel('Операционная система')
plt.ylabel('Количество ноутбуков')
plt.xticks(rotation=45)
;
```

### Постройте матрицу корреляции

```python
corr_matrix = df.corr()
corr_matrix
```

* до 0,2 - Очень слабая
* до 0,5 - Слабая
* до 0,7 - Средняя
* до 0,9 - Высокая
* свыше 0,9 - Очень высокая корреляция

```python
corr_matrix[np.abs(corr_matrix)<=0.2] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True);
```

```python
upp_mat = np.triu(df.corr())
```

```python
sns.heatmap(df.corr(), annot=True, cmap='viridis', mask=upp_mat);
```

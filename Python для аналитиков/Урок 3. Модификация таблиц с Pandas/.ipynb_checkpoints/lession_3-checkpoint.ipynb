```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Churn_Modelling.csv')
df
```

```python
bank = df[['CustomerId','CreditScore','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','Exited']]
```

```python
users = df[['CustomerId','Surname','Geography','Gender','Age','EstimatedSalary']]
```

```python
users.head()
```

```python
users.shape
```

### Создание новых признаков

```python
users['new_feature'] = 0
users.head()
```

**1 способ. Просто добавить новый столбец**

```python
users['Age(days)'] = users['Age']*365
users.head()
```

**2 способ. Перебором через цикл записать нужные данные в список, а потом добавить в столбец**

```python
age_days = []
for i, row in users.iterrows():
    age_days.append(row['Age']*365)
    
users['Age(days)_2'] = age_days
users.head()
```

**3 способ. Через метод .apply, который принимает в себя функцию**

```python
def age_to_days(x):
    return x*365

users['Age(days)_3'] = users['Age'].apply(age_to_days)
users.head()
```

**В методе .apply можно наблюдать сколько он уже прошел строчек и приблизительное время ожидания. Все это делается с помощью библиотеки tqdm**

```python
import time
from tqdm import tqdm
tqdm.pandas()


def age_to_days(x):
    #time.sleep(0.1e-100)  # задержка времени, чтобы визуально оценить эффект
    return x * 365

users['Age'].progress_apply(age_to_days)
```

```python
users = users.drop(columns='new_feature')
users.head()
```

```python
users['new_feature'] = 0
users.drop(columns='new_feature', inplace=True)
users.head()
```

**Одновременно можно удалить сразу несколько колонок**

```python
users.drop(columns=['Age(days)', 'Age(days)_2', 'Age(days)_3'], inplace=True)
users.head()
```

### Методы агрегации

Агрегация - процесс объединения элементов в одно значение\
Первый способ: вызвать метод **.agg()** у pd.Series и передать в него список желаемых агрегаций

```python
users['Age'].agg(['min', 'max'])
```

```python
users.agg({
    'Age': ['min', 'max'],
    'EstimatedSalary': 'mean'
})
```

```python
users.agg(
    min_age=('Age', 'min'),
    max_age=('Age', 'max'),
    mean_salary=('EstimatedSalary', 'mean')
)
```

**Атрибут how()** - присутствует в обеих методах

<img src='1550660883.jpg'>

**left**

```python
merged_left = bank.merge(users, on='CustomerId', how='left')
merged_left.shape
```

проверка на слияние. если 0, то все данные заполнились. Если будут пропуски, то их можно отдельно посмотреть

```python
merged_left.isna().sum()
```

В объединенный Df передаем маску проверки на 0 и получаем элементы, которые были пропущены

```python
merged_left[merged_left['Age'].isna()] # смотрим, если пропуски есть
```

**right**

```python
merged_right = bank.merge(users, on='CustomerId', how='right')
merged_right.shape
```

```python
merged_right.isna().sum()
```

**inner**

```python
merged_inner = bank.merge(users, on='CustomerId', how='inner')
merged_inner.shape
```

```python
merged_inner.isna().sum()
```

**outer**

```python
merged_outer = bank.merge(users, on='CustomerId', how='outer')
merged_outer.shape
```

### Методы группировок

Используется для объединения по нескольким признакам

**groupby**
<img src='group2.webp'>

```python
toy_df = pd.DataFrame({
    'client_id': [1, 2, 2, 3, 1, 1],
    'item': ['chocolate', 'cheese', 'ham', 'candy', 'chair', 'book'],
    'price': [68, 280, 302, 39, 2099, 1089]
})

toy_df
```

```python
grouped = toy_df.groupby('client_id')
grouped
```

```python
grouped.groups
```

```python
grouped.sum()
```

```python
grouped.agg({'price': ['sum', 'min', 'max']})
```

```python
users.groupby('Geography').agg({'Age': ['mean'], 'EstimatedSalary': ['min']})
```

**pivot_table**


```python
toy_df
```

```python
toy_df.pivot_table(index='client_id',
                   values='price',
                   aggfunc='sum')
```

```python
users.pivot_table(index='Geography',
                  aggfunc={'Age': ['mean'], 'EstimatedSalary': 'min'})
```

```python
users.pivot_table(index='Geography',  # строка
                  columns='Gender',   # столбцы
                  values='EstimatedSalary',  # какой признак берем
                  aggfunc='mean',  # что с ним делаем
                  margins=True,    # колонка "итого" (по столбам и строкам)
                  margins_name='Total')  # и ее название
```

**crosstable**


встроенная функция в pandas, которая принимает работает со всем df и принимает не названия столбцов, а сам столбец, то есть pd.Series

```python
pd.crosstab(index=users['Geography'],  # значение по строкам
            columns=users['Gender'])   # значение по столбцам
```

```python
pd.crosstab(index=users['Geography'],
            columns=users['Gender'],
            values=users['EstimatedSalary'],  # стоблец с которым работаем
            aggfunc='mean')  # функцию которую применяем
```

```python
pd.crosstab(index=users['Geography'],
            columns=users['Gender'],
            normalize='all') # доля по всей таблице
```

```python
pd.crosstab(index=users['Geography'],
            columns=users['Gender'],
            normalize='index')  # доля по строчкам (в сумме 100%)
```

```python
pd.crosstab(index=users['Geography'],
            columns=users['Gender'],
            normalize='columns')  # доля по колонкам( в сумме 100%)
```

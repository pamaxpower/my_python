```python
df = pd.read_csv('./kc_house_data.csv', encoding='latin-1')
df.head()
```

```python
df.info()
```

### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df['price_per_sq_lot'] = df['price']/df['sqft_lot']
df
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: x if x>0 else 0)
df[['delta_renovated', 'yr_renovated', 'yr_built']].sample(10)
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_sale'] = df['date'].apply(lambda x: int(x[:4]))
df['month_sale'] = df['date'].apply(lambda x: int(x[4:6]))
df[['date','year_sale','month_sale']]
                                   
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(['date', 'zipcode', 'lat', 'long'], axis=1, inplace=True)
df
```

## Задача 3
### Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
# df.groupby('price').agg({'bedrooms': 'mean'}).sort_values('price')
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
group_price = df.groupby('condition').agg({'price': ['max', 'min', 'mean']})
group_price
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(
    index = 'waterfront',
    columns = 'view',
    aggfunc = 'count',
    values = 'id',
    fill_value = '--'
)
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
def highligth_max(x):
    return np.where(x == np.max(x.to_numpy()), f"color: red ;", None)
```

```python
house = pd.crosstab(
    index = df['floors'],
    columns = df['bedrooms']
)
house.style.apply(highligth_max, axis = None)
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.pivot_table(
    index = 'grade',
    columns = 'condition',
    aggfunc = 'median',
    values = 'price',
    fill_value = '--'
)
```

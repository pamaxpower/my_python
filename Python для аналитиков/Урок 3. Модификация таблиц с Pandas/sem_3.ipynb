### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu'].value_counts()
```

Название фирмы находится в первом слове значения столбца Cpu

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df['Cpu_Company'].value_counts()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()
```

Как видим в значениях памяти разные наименования(GB и TB). Для этой решения задачи нужно написать функцию, которая отделить циферное обозначение памяти от букв, и переведет TB в GB. \
отделить числовую часть от буквенной можно двумя способами. Причем значения с TB будет 1.0, что можно перевести в int только через обертку float

```python
def gb(s):
    s = s.split()[0]
    if 'TB' in s:
        x = int(float(s.replace('TB',''))) * 1024
    elif s.endswith('GB'):
        x = int(s.replace('GB',''))
    return x


df['Memory_Amount'] = df['Memory'].apply(gb)
df['Memory_Amount'].value_counts()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
    df['Memory_Type'] = df['Memory'].apply(lambda x: x.split(' ')[-1])
    df['Memory_Type'].value_counts()
```

### 5.2 С каким типом памяти и с каким объемом памяти больше ноутбуков?

```python
data = pd.crosstab(
    index = df['Memory_Type'],
    columns = df['Memory_Amount']
)
```

```python
def gold_max(x):
    m = x.to_numpy().max()
    color = {True: "background-color: #c78f2e", False: ""}
    is_max = (x == m).replace(color)
    return is_max
```

```python
data.style.apply(gold_max, axis = None)
```

```python
data.to_numpy() # делает массив из датафрейма
```

```python
def highligth_max(x):
    return np.where(x ==np.max(x.to_numpy()), f"color: red ;", None)
```

```python
data.style.apply(highligth_max, axis = 0) # по столбцам
```

```python
data.style.apply(highligth_max, axis = 1) # по строкам
```

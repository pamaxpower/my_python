# Задание 5.

### 5.1. Выберите клиенту ноутбук. Клиент хочет подобрать ноутбук с 8GB или 16GB ОЗУ на Windows 10 в стоимости до 500 евро, сколько у него вариантов?

```python
df[
    (df['TypeName'] == 'Notebook') &
   ((df['Ram'] == '8GB') | (df['Ram'] == '16GB')) &
   (df['OpSys'] == 'Windows 10') &
   (df['Price_euros'] < 500)
].count()
```

```python
df[
    (df['TypeName'] == 'Notebook') &
    (df['Ram'].isin(['8GB','16GB'])) &
   (df['OpSys'] == 'Windows 10') &
   (df['Price_euros'] < 500)
].shape[0]
```

### 5.2. Выберите клиенту ноутбук. Клиент хочет подобрать ноутбук от MSI, с видеокартой Nvidia GeForce GTX 1050 Ti и главное не с диагональю 15,6. В какой ценовой категории вышли подобные ноутбуки?

```python
sort1 = df[
    (df['Company'] == 'MSI') &
    (df['Gpu'] == 'Nvidia GeForce GTX 1050 Ti') &
    ~(df['Inches'] == 15.6)
]['Price_euros']
sort1.min(), sort1.max()
```

### 5.3 Что дешевле? Ноутбуки с CPU Intel Core i7 7700HQ 2.8GHz или с Intel Core i7 7600U 2.8GHz?

```python
n1 = df[
    df['Cpu']=='Intel Core i7 7700HQ 2.8GHz'
]['Price_euros'].mean()
n2 = df[
    df['Cpu']=='Intel Core i7 7600U 2.8GHz'
]['Price_euros'].mean()

n1,n2
```
